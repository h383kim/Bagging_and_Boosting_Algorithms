```python
import pandas as pd
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import AdaBoostClassifier
```

# Importing dataset (sklearn Breast Cancer dataset)

```python
# Loading data
data = load_breast_cancer()
# Making features dataframe
data_df = pd.DataFrame(data=data.data, columns=data.feature_names)
# Making label dataframe
target_df = pd.DataFrame({"Target": data.target})
```

```python
data_df
```

```python
target_df
```

# Train & Predict Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split train_test dataset
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Model Instance
clf = AdaBoostClassifier(n_estimators=50, random_state=42)
# Train(fit) the model to the train set
clf.fit(X_train, y_train)
# Predict with test set
y_pred = clf.predict(X_test)

# See results
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

# Evaluation Metrics

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Access one of the estimators (e.g., the first one)
estimator = clf.estimators_[0]

# Plot one of the decision trees(estimators)
plt.figure(figsize=(12, 8))
plot_tree(estimator,
          feature_names=data.feature_names,
          class_names=clf.classes_.astype(str),
          filled=True,
          rounded=True)
plt.title('Decision Tree in AdaBoost Ensemble')
plt.show()
```

```python
# Saving the importance of each stump(weak learner)
feature_importances = pd.DataFrame({
    'Feature': data.feature_names,
    'Importance': clf.feature_importances_
})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Plot the bar graph
plt.figure(figsize=(8, 6))
plt.bar(feature_importances['Feature'], feature_importances['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.show()
```

```python
weights = clf.estimator_weights_
print(f"Weights of each weak learner: {weights}")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title('Confusion Matrix for AdaBoost Classifier')
plt.show()
```
